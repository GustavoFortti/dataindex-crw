```python
from functions import get_all_origins, WORDLIST, clean_text, best_words
import re
import math
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)


WORDLIST = WORDLIST["supplement"]
wordlist = [item for sublist in WORDLIST for item in sublist]

df = get_all_origins()
# df[["title", "spec", "spec_route"]]
```

```python
word_count = {}  # Dicionrio para armazenar a contagem para cada palavra

for word in wordlist:
    word_count[word] = {}
    df_temp = df[df["spec_route"].str.contains(word, case=False, na=False)]
    for word_2 in wordlist:
        count = df_temp["spec_route"].str.contains(word_2, case=False, na=False).sum()
        word_count[word][word_2] = count
    
    # Ordena o dicionrio interno para a palavra atual 'word' pelo valor de contagem
    word_count[word] = dict(sorted(word_count[word].items(), key=lambda item: item[1], reverse=True))

print(word_count)
word_count = pd.DataFrame(word_count)
```
